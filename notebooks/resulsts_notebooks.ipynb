```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ROLLING_WINDOW = 500

# Load the total_rewards from the pickle file
with open("../checkpoints/DDQN/total_rewards.pkl", 'rb') as f:
    total_rewards = pickle.load(f)

# Calculate the smoothed rewards over a window of 10 episodes
smoothed_rewards = pd.Series(total_rewards).rolling(window=ROLLING_WINDOW).mean()

# Plot the smoothed rewards
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(total_rewards) + 1), smoothed_rewards)

plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Mean Reward per Episode')
plt.legend()
plt.show()
```
