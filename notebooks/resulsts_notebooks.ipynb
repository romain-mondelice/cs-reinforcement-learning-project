```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ROLLING_WINDOW = 500

# Load the total_rewards from the pickle file
with open("../checkpoints/DQN/total_rewards.pkl", 'rb') as f:
    total_rewards_dqn = pickle.load(f)

# Calculate the smoothed rewards over a window of 10 episodes
smoothed_rewards_dqn = pd.Series(total_rewards_dqn).rolling(window=ROLLING_WINDOW).mean()

# Plot the smoothed rewards
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(total_rewards_dqn) + 1), smoothed_rewards_dqn)

plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Mean Reward per Episode')
plt.legend()
plt.show()
```

```python
ROLLING_WINDOW = 500

# Load the total_rewards from the pickle file
with open("../checkpoints/DDQN/total_rewards.pkl", 'rb') as f:
    total_rewards_ddqn = pickle.load(f)

# Calculate the smoothed rewards over a window of 10 episodes
smoothed_rewards_ddqn = pd.Series(total_rewards_ddqn).rolling(window=ROLLING_WINDOW).mean()

# Plot the smoothed rewards
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(total_rewards_ddqn) + 1), smoothed_rewards_ddqn)

plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Mean Reward per Episode')
plt.legend()
plt.show()
```

```python
# Plot the smoothed rewards for both DQN and DDQN
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(total_rewards_dqn) + 1), smoothed_rewards_dqn, label='DQN')
plt.plot(range(1, len(total_rewards_ddqn) + 1), smoothed_rewards_ddqn, label='DDQN')

plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Mean Reward per Episode for DQN and DDQN')
plt.legend()
plt.show()
```
